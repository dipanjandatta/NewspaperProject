```
# prompt: download an image from a link and show that image

import requests
from PIL import Image
from io import BytesIO

image_url = "https://images.eap.bl.uk/EAP262/EAP262_1_1_45_269/1.jp2/full/6854,/0/default.jpg" #@param {type:"string"}

response = requests.get(image_url)
img = Image.open(BytesIO(response.content))

display(img)
```
